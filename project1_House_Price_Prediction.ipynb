```python
# Impoort required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import pandas as pd
df = pd.read_csv('house_prices.csv')
print (df.head())
```

```python
# Visualize the data
import matplotlib.pyplot as plt

plt.scatter(df['Size (sq ft)'], df['Price ($)'], color='Blue')
plt.title('House vs Price')
plt.xlabel('Size (sq ft)')
plt.ylabel('price ($)')
plt.grid(True)
plt.show()
```

```python
# Split [the dataset(train-test)
x = df[['Size (sq ft)']]
y = df[['Price ($)']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# Train a simple linear regressioon model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# Make predictions and evaluate test data
y_pred = model.predict(x_test)

# Evaluation metrics(model performance)
r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)

print(f"R2 score: {r2}")
print(f"Mean Squared Error: {mse}")
```

```python
# Visualize the regression line

# plot training data and regression
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, model.predict(x), color='red', label='Regression line')
plt.title('Linear Regressioon - House Price Prediction')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
